### Kdeplot


A KDE plot, or kernel density estimate plot, is a type of data visualization that provides a smooth estimate of the probability density function of a random variable. The purpose of a KDE plot is to estimate the underlying distribution of a set of data, and to provide a visual representation of that distribution.

KDE plots are used for a variety of purposes, including:

- Data exploration: KDE plots can be used to quickly visualize the distribution of a set of data, providing insights into the underlying distribution of the data.

- Data analysis: KDE plots can be used to identify the presence of outliers, skewness, or multimodality in the data.

- Data representation: KDE plots can be used to represent data in a way that is easy to understand and visually appealing, making it useful for presentations and reports.

- Data comparison: KDE plots can be used to compare two or more sets of data, allowing users to quickly identify differences and similarities in the distribution of the data.

- Statistical modeling: KDE plots can be used as a tool for statistical modeling, as they provide a smooth estimate of the underlying distribution of the data, which can be useful for building predictive models.

Overall, the purpose of a KDE plot is to provide a smooth estimate of the underlying distribution of a set of data that is easy to understand and interpret, and that can be used to support data exploration, analysis, and decision-making.

```python
# Kdeplot -  kernel denstity estimation -  Density of probablity in data 
sns.kdeplot(data = tips,x = 'total_bill')
```

```python
sns.displot(data = tips,x = 'total_bill',kind = 'kde')
```

```python
# hue -> fill
sns.displot(data = tips,x = 'total_bill',kind = 'kde',hue = 'sex',fill = True)
```

### Rugplot

A rug plot, also known as a marginal plot, is a type of data visualization that provides a graphical representation of the distribution of a single variable. The purpose of a rug plot is to provide a visual representation of the density and distribution of the data, and to give a sense of the distribution of individual observations.

Rug plots are used for a variety of purposes, including:

- Data exploration: Rug plots can be used to quickly visualize the distribution of a single variable, providing insights into the distribution of the data.

- Data analysis: Rug plots can be used to identify the presence of outliers or skewness in the data.

- Data representation: Rug plots can be used to represent data in a way that is easy to understand and visually appealing, making it useful for presentations and reports.

- Data comparison: Rug plots can be used to compare two or more sets of data, allowing users to quickly identify differences and similarities in the distribution of the data.

Overall, the purpose of a rug plot is to provide a visual representation of the distribution of a single variable that is easy to understand and interpret, and that can be used to support data exploration, analysis, and decision-making.

```python
#Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.



sns.kdeplot(data = tips,x = 'total_bill')
sns.rugplot(data = tips,x = 'total_bill')
```

### HeatMap
A heatmap is a graphical representation of data that uses a color-coding system to represent different values. The purpose of a heatmap is to provide a visual representation of data that can be easily understood and interpreted.

Heatmaps are used for a variety of purposes, including:

- Data exploration: Heatmaps can be used to quickly visualize the distribution and patterns in large datasets.

- Data analysis: Heatmaps can be used to identify correlations and relationships between variables in the data.

- Outlier detection: Heatmaps can be used to quickly identify outliers, or observations that are significantly different from the rest of the data.

- Data representation: Heatmaps can be used to represent data in a way that is easy to understand and visually appealing, making it useful for presentations and reports.

- Data comparison: Heatmaps can be used to compare two or more sets of data, allowing users to quickly identify differences and similarities.

Overall, the purpose of a heatmap is to provide a visual representation of data that is easy to understand and interpret, and that can be used to support data exploration, analysis, and decision-making.

```python
#Heatmap

#Plot rectangular data as a color encoded matrix
temp_df = gap.pivot(index = 'country', columns = 'year',values='lifeExp')


# axes level funcion
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

Clustermaps can be used for a variety of purposes, such as:

- Outlier detection: Outliers, or observations that are significantly different from the rest of the data, can be easily identified in a clustermap.

- Data exploration: Clustermaps can be used to explore the relationships between variables and to identify patterns in the data.

- Data analysis: Clustermaps can be used as a preprocessing step in machine learning and other data analysis techniques, as they can help to identify correlations and relationships in the data that can be useful in building predictive models.

- Overall, the purpose of a clustermap is to help data scientists and analysts better understand their data and make informed decisions based on that understanding.

### Box Plot

A box plot is a graphical representation of a dataset that displays the median, quartiles, and outliers of the data. The box represents the middle 50% of the data, with the lower edge of the box at the first quartile, the upper edge at the third quartile, and a line inside the box representing the median. The "whiskers" extend to the lowest and highest data points that are within 1.5 times the interquartile range (the distance between the first and third quartiles), and any data points outside this range are plotted as individual points or "outliers".

The purpose of a box plot is to provide a visual summary of the distribution of a dataset, including its central tendency, spread, and any outliers. It can be particularly useful for comparing the distributions of multiple datasets or groups, and for identifying any differences in their medians, quartiles, or ranges.

```python
# Box Plot
sns.boxplot(data = tips,x = 'sex',y='total_bill')
```

```python
sns.catplot(data = tips,x = 'sex',y='total_bill',kind = 'box',hue = 'day')
```

```python
#single box plot -> numerical col
sns.boxplot(data = tips, y= 'total_bill')
```

### violinplot = (boxplot + KdePlot)

A violin plot is a graphical representation of a dataset that combines the features of a box plot and a kernel density plot. It shows the same information as a box plot, with the addition of a density curve on each side, which provides a more detailed picture of the distribution of the data.

The shape of the density curve reflects the density of the data at different values, with wider sections indicating more data points and narrower sections indicating fewer data points. The wider sections are "fatter" and the narrower sections are "thinner". The kernel density plot can also help to identify multiple peaks or modes in the distribution.

The purpose of a violin plot is similar to that of a box plot, to provide a visual summary of the distribution of a dataset, including its central tendency, spread, and any outliers. However, the violin plot provides more information about the shape and density of the distribution, which can be particularly useful when comparing multiple datasets or groups. The violin plot is often used in exploratory data analysis and statistical inference to quickly understand the distribution of the data and to make informed decisions about further analysis.

```python
# violionplot 
sns.violinplot(data = tips,x='day',y='total_bill')
```

```python
sns.catplot(data = tips,x='day',y='total_bill',kind = 'violin',hue = 'sex',split = True)
```

# Barplot

A categorical bar plot is a type of plot in Seaborn that allows you to visualize the distribution of a categorical variable and its corresponding values in a dataset. It is often used to show the relationship between a categorical variable and a numerical variable.

To create a categorical bar plot in Seaborn, you can use the sns.barplot() function. This function takes in a categorical variable as the x-axis and a numerical variable as the y-axis. It then calculates the summary statistics for each category of the x-axis variable and plots the corresponding bar heights.

The bar heights represent the mean or median of the numerical variable for each category, and the error bars represent the standard deviation or confidence interval. This allows you to see the overall trend of the data and the variability within each category.

Categorical bar plots can be further customized using Seaborn's styling options. For example, you can change the color and style of the bars and error bars, add labels to the axes and bars, and adjust the size and aspect ratio of the plot.

Overall, categorical bar plots are a simple yet effective way to explore the distribution of a categorical variable and its relationship with a numerical variable in a dataset.

```python
import numpy as np
sns.barplot(data = tips, x = 'sex', y='total_bill',ci=None,hue = 'smoker')
```

```python
sns.catplot(data = tips, x = 'sex', y='total_bill',kind = 'bar',ci=None)
```

# Point Plot
A categorical point plot is a type of plot in Seaborn that shows the relationship between a categorical variable and a numerical variable by plotting the point estimates and confidence intervals of the means or medians of the data.

To create a categorical point plot in Seaborn, you can use the sns.pointplot() function, which takes in a categorical variable as the x-axis and a numerical variable as the y-axis. It then calculates the summary statistics for each category of the x-axis variable and plots the corresponding points at the mean or median of the y-axis variable, along with error bars that represent the confidence interval.

The advantage of using a point plot is that it allows you to compare the mean or median values between different categories in a visually appealing way. Additionally, you can use hue or style attributes to show a third categorical variable in the plot.

Categorical point plots can be customized using Seaborn's styling options, allowing you to adjust the size, aspect ratio, and color of the plot, as well as the font and labels of the axes and legend.

Overall, categorical point plots are a useful tool for visualizing the relationship between a categorical variable and a numerical variable in a dataset, and can be used for exploratory data analysis or for communicating insights in a clear and concise manner.


```python
#point plot

sns.pointplot(data = tips, x = 'sex', y='total_bill',hue = 'smoker',ci = None)
```

# Count Plot
A categorical count plot is a type of plot in Seaborn that shows the number of occurrences of each category in a categorical variable.

To create a categorical count plot in Seaborn, you can use the sns.countplot() function, which takes in a categorical variable as the x-axis and plots the count of each category. The counts are represented as vertical bars, with the height of each bar corresponding to the number of observations in each category.

The categorical count plot is useful for visualizing the frequency of occurrence of each category in a categorical variable, and can help to identify any imbalances or biases in the data. It can also be used to compare the relative frequencies of different categories.

Categorical count plots can be customized using Seaborn's styling options, allowing you to adjust the size, aspect ratio, and color of the plot, as well as the font and labels of the axes.

Overall, categorical count plots are a simple yet effective tool for exploring the distribution of a categorical variable in a dataset, and can be used as a first step in data analysis or for communicating insights in a clear and concise manner.

```python
# CountPlot
sns.countplot(data = tips,x = 'sex',hue = 'day')
```

###  Facet Grid

Facet Grid is a plotting tool in the Seaborn data visualization library that allows you to create a grid of subplots based on unique combinations of one or more categorical variables in a dataset.

The purpose of Facet Grid is to enable you to create multiple plots that share a common structure but display different subsets of the data, based on the values of one or more categorical variables. This can be particularly useful for exploring complex relationships between multiple variables or for visualizing trends across different categories of data.

With Facet Grid, you can specify the variables to use for row and column facets, as well as additional facets for each subplot. You can also specify the type of plot to create for each subplot, such as scatter plots, line plots, or bar plots, and customize the appearance of each subplot using Seaborn's wide range of styling options.

Overall, Facet Grid is a powerful tool for creating multi-panel visualizations that allow you to explore the relationships and patterns within your data in a more comprehensive and structured way.

```python
g = sns.FacetGrid(data=tips,col='day',row='time',hue = 'smoker')
g.map(sns.scatterplot,'tip','total_bill')
g.add_legend()
```

### Pairwise relationship - PairPolt

```python
iris = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']]
```

```python
sns.pairplot(iris,hue = 'species')
```

```python
g = sns.PairGrid(data = iris,hue = 'species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data = iris,hue = 'species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data = iris,hue = 'species',vars=['sepal_length','sepal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
