### Kdeplot


A KDE plot, or kernel density estimate plot, is a type of data visualization that provides a smooth estimate of the probability density function of a random variable. The purpose of a KDE plot is to estimate the underlying distribution of a set of data, and to provide a visual representation of that distribution.

KDE plots are used for a variety of purposes, including:

- Data exploration: KDE plots can be used to quickly visualize the distribution of a set of data, providing insights into the underlying distribution of the data.

- Data analysis: KDE plots can be used to identify the presence of outliers, skewness, or multimodality in the data.

- Data representation: KDE plots can be used to represent data in a way that is easy to understand and visually appealing, making it useful for presentations and reports.

- Data comparison: KDE plots can be used to compare two or more sets of data, allowing users to quickly identify differences and similarities in the distribution of the data.

- Statistical modeling: KDE plots can be used as a tool for statistical modeling, as they provide a smooth estimate of the underlying distribution of the data, which can be useful for building predictive models.

Overall, the purpose of a KDE plot is to provide a smooth estimate of the underlying distribution of a set of data that is easy to understand and interpret, and that can be used to support data exploration, analysis, and decision-making.

```python
# Kdeplot -  kernel denstity estimation -  Density of probablity in data 
sns.kdeplot(data = tips,x = 'total_bill')
```

```python
sns.displot(data = tips,x = 'total_bill',kind = 'kde')
```

```python
# hue -> fill
sns.displot(data = tips,x = 'total_bill',kind = 'kde',hue = 'sex',fill = True)
```

### Rugplot

A rug plot, also known as a marginal plot, is a type of data visualization that provides a graphical representation of the distribution of a single variable. The purpose of a rug plot is to provide a visual representation of the density and distribution of the data, and to give a sense of the distribution of individual observations.

Rug plots are used for a variety of purposes, including:

- Data exploration: Rug plots can be used to quickly visualize the distribution of a single variable, providing insights into the distribution of the data.

- Data analysis: Rug plots can be used to identify the presence of outliers or skewness in the data.

- Data representation: Rug plots can be used to represent data in a way that is easy to understand and visually appealing, making it useful for presentations and reports.

- Data comparison: Rug plots can be used to compare two or more sets of data, allowing users to quickly identify differences and similarities in the distribution of the data.

Overall, the purpose of a rug plot is to provide a visual representation of the distribution of a single variable that is easy to understand and interpret, and that can be used to support data exploration, analysis, and decision-making.

```python
#Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.



sns.kdeplot(data = tips,x = 'total_bill')
sns.rugplot(data = tips,x = 'total_bill')
```

```python
#bivarite Histogram
# sns.histplot(data = tips, x = 'total_bill',y='tip')
sns.displot(data = tips, x = 'total_bill',y='tip',kind ='hist')
```

```python
#bivariate kdeplot
sns.kdeplot(data = tips, x = 'total_bill',y='tip')
```

### HeatMap
A heatmap is a graphical representation of data that uses a color-coding system to represent different values. The purpose of a heatmap is to provide a visual representation of data that can be easily understood and interpreted.

Heatmaps are used for a variety of purposes, including:

- Data exploration: Heatmaps can be used to quickly visualize the distribution and patterns in large datasets.

- Data analysis: Heatmaps can be used to identify correlations and relationships between variables in the data.

- Outlier detection: Heatmaps can be used to quickly identify outliers, or observations that are significantly different from the rest of the data.

- Data representation: Heatmaps can be used to represent data in a way that is easy to understand and visually appealing, making it useful for presentations and reports.

- Data comparison: Heatmaps can be used to compare two or more sets of data, allowing users to quickly identify differences and similarities.

Overall, the purpose of a heatmap is to provide a visual representation of data that is easy to understand and interpret, and that can be used to support data exploration, analysis, and decision-making.

```python
#Heatmap

#Plot rectangular data as a color encoded matrix
temp_df = gap.pivot(index = 'country', columns = 'year',values='lifeExp')


# axes level funcion
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot
temp_df = gap[gap['continent']=='Europe'].pivot(index = 'country', columns = 'year',values='lifeExp')
# axes level funcion
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot = True,linewidth = 0.5,cmap = 'viridis')
```

### ClusterMap
A clustermap is a graphical representation of a heatmap in which similar rows or columns are grouped together (clustered) based on their similarity. The purpose of a clustermap is to visualize and explore the structure of high-dimensional data, and to identify patterns and relationships between the observations and variables.

Clustermaps can be used for a variety of purposes, such as:

- Dimensionality reduction: By clustering similar rows or columns, a clustermap can provide a lower-dimensional representation of the data, making it easier to visualize and understand.

- Outlier detection: Outliers, or observations that are significantly different from the rest of the data, can be easily identified in a clustermap.

- Data exploration: Clustermaps can be used to explore the relationships between variables and to identify patterns in the data.

- Data analysis: Clustermaps can be used as a preprocessing step in machine learning and other data analysis techniques, as they can help to identify correlations and relationships in the data that can be useful in building predictive models.

- Overall, the purpose of a clustermap is to help data scientists and analysts better understand their data and make informed decisions based on that understanding.



```python
#clustermap
iris = px.data.iris()
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```

###  Facet Grid

```python
g = sns.FacetGrid(data=tips,col='day',row='time',hue = 'smoker')
g.map(sns.scatterplot,'tip','total_bill')
g.add_legend()
```

### Pairwise relationship - PairPolt

```python
sns.pairplot(iris,hue = 'species')
```

```python
g = sns.PairGrid(data = iris,hue = 'species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data = iris,hue = 'species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data = iris,hue = 'species',vars=['sepal_length','sepal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

### JointPlot vs JointGrid

```python
g = sns.JointGrid(data = tips,x = 'total_bill',y='tip',hue = 'sex')
g.plot(sns.scatterplot,sns.boxplot)
```
