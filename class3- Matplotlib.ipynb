# 2D Line Plot

- A 2D line plot, also known as a line graph or a line chart, is a visual representation of data that uses points connected by lines to display changes over a certain time period or for multiple categories.

- Bivariat Analysis :
Bivariate analysis is a statistical method that helps analyze the relationship between two variables. It involves examining the relationship between two variables and determining if a relationship exists between them and what kind of relationship exists. It can be performed using graphical methods such as scatter plots and correlation coefficients, or through more advanced statistical techniques such as regression analysis. Bivariate analysis can provide insights into the nature of the relationship between two variables, including the strength, direction, and form of the relationship, which can inform further research and decision-making.
- Categorical->Numerical and Numrical -> Numerical
## Use case
- Time-series data: Line plots are often used to display changes in data over time, such as stock prices, weather patterns, or sales data.

- Comparison of multiple categories: Line plots can be used to compare multiple sets of data, such as the performance of different stocks or the growth of multiple cities.

- Trends and patterns: Line plots can be used to identify trends and patterns in data, such as increasing or decreasing trends, seasonal patterns, and fluctuations.

- Presenting data in a clear and concise manner: Line plots are a simple and effective way to present data to others, especially for data with multiple categories or time-series data.

# Scatter Plot 

- A scatter plot, also known as a scatter graph or scatter chart, is a type of plot used to display the relationship between two variables. It plots individual data points as points on a 2D graph and can be used to identify patterns and correlations.


- Bivariate Analysis
- Numerical vs Numerical

### Use Case 

- Correlation analysis: Scatter plots are often used to determine the correlation between two variables, such as the relationship between height and weight, or the relationship between the number of hours studied and exam scores.

- Outlier detection: Scatter plots can help identify outliers, which are data points that are significantly different from the rest of the data.

- Comparison of multiple data sets: Scatter plots can be used to compare multiple data sets and visualize the differences and similarities between them.

- Function fitting: Scatter plots can be used to fit a function to a set of data, such as a linear regression model or a polynomial function.

- Multivariate data visualization: Scatter plots can be used to visualize multivariate data, where multiple variables are present, and help to identify patterns and correlations between the variables.

```python
#plt.scatter Simple Function
x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

# Bar Chart

- A bar chart, also known as a bar graph or bar diagram, is a type of chart used to display categorical data. It consists of bars of different heights or lengths that represent the values of each category.

- Bivariate Analysis
- Numerical Vs Categorical

### Use Case


- Comparison of categorical data: Bar charts are often used to compare categorical data, such as the number of products sold in different regions, the number of votes received by different political parties, or the popularity of different brands.

- Distribution of data: Bar charts can be used to display the distribution of data, such as the distribution of ages in a population or the distribution of grades in a class.

- Presenting data in a clear and concise manner: Bar charts are a simple and effective way to present data to others, especially for data with multiple categories.

- Displaying changes over time: Bar charts can be used to display changes in data over time, such as the changes in sales of a product over several months.

- Displaying data with discrete categories: Bar charts are particularly useful for displaying data with discrete categories, such as the number of students in different majors or the number of books in different genres.

```python
#simple bar chart
kids = [10,20,40,10,30]
cartoon = ['Bugs Bunny','The Simpsons','Tom & Jerry','Mickey Mouse','Scooby-Doo']
plt.bar(cartoon,kids,color = 'green')
```

```python
#horizontal bar chart More - no of Categroies are more 
plt.barh(cartoon,kids,color = 'green')
```

```python
#One common Problem 
kids = [10,20,40,10,30]
cartoon = ['Bugs Bunny','The Simpsons','Tom and Jerry','Mickey Mouse','Scooby-Doo']
plt.bar(cartoon,kids,color = 'green')
plt.xticks(rotation = "vertical")
plt.show()
```

```python
# Stacked bar Chart
plt.bar(top_scorer['Player'],top_scorer['2019'],label = '2019')
plt.bar(top_scorer['Player'],top_scorer['2021'],bottom = top_scorer['2019'],label = '2021')
plt.bar(top_scorer['Player'],top_scorer['2022'],bottom = (top_scorer['2021'] + top_scorer['2019']),label = '2022')

plt.legend()
plt.show()
```

# Histogram

- A histogram is a graphical representation of the distribution of a set of continuous data. It is an estimate of the probability distribution of a continuous variable. The data is divided into a set of bins and the height of each bar in the histogram represents the frequency of data within the corresponding bin.

- univariate Analysis 
- Numerical col
### Use case
- Data distribution analysis: Histograms are often used to visualize the distribution of a dataset and identify the shape of the distribution, such as symmetrical, skewed, or multimodal.

- Outlier detection: Histograms can help identify outliers, which are data points that are significantly different from the rest of the data.

- Comparison of data sets: Histograms can be used to compare multiple data sets and visualize the differences and similarities between them.

- Identifying the range and density of data: Histograms can be used to identify the range and density of data and help to determine if the data is centered around a particular value.

- Quality control: Histograms are used in quality control to ensure that a process is producing outputs within specified limits.

```python
data = [32,35,30,10,15,27,10]

plt.hist(data,bins=[10,15,20,25,30,35,40])
```

# pie charts

- A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. Each slice represents a category and the size of the slice is proportional to the quantity it represents.

- Univariate/Bivariate Analysis
- categorical Vs Numerical
### Use Case  

- Proportion representation: Pie charts are used to represent the proportion of different categories in a whole. For example, a pie chart can be used to show the proportion of expenses in a budget for different categories.

- Simple data comparison: Pie charts are useful for comparing simple data sets with only a few categories, where the relationship between categories can be easily understood.

- Showing parts-of-a-whole: Pie charts can be used to show how a single data point is divided into smaller parts, such as the distribution of a company's profits among different departments.

- Visualizing relative proportions: Pie charts can be used to visualize the relative proportions of different categories, such as the distribution of market share among different brands.

- Presenting data in a clear and concise manner: Pie charts are a simple and effective way to present data to others, especially when there are only a few categories and the data is easy to understand.

```python
# Simple Data
data  = [45,49,100,20,49]
subjects = ['eng','science','maths','Social Science','Hindi']
plt.pie(data,labels = subjects)
plt.show()
```
